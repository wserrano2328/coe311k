```python
import numpy
import matplotlib.pyplot as plt
```

```python
def approxPi(numSamples):
    counter = 0
    for s in range(numSamples):
        x = numpy.random.uniform(0, 2) - 1
        y = numpy.random.uniform(0, 2) - 1
        
        if (x**2 + y**2 < 1):
            counter= counter + 1
            
    return 4.0 * counter / numSamples
```

```python
errorSet = []
realPi = numpy.pi
numSampleSet = numpy.arange(10, 10030, 20)
for i in numSampleSet:
    calculatedPi = approxPi(i)
    error = abs(calculatedPi - realPi)
    errorSet = numpy.append(errorSet, error)
```

```python
plt.plot(numSampleSet, errorSet, 'r')
plt.xlabel("Number of samples in Monte Carlo")
plt.ylabel("Error")
plt.show()
```
