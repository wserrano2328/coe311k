```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def taylorSin(x, terms):
    approxSin = 0
    for i in range (0, terms):
        power = 2*i + 1
        term = (-1)**i * (x**power) / math.factorial(power)
        approxSin += term
        
    return approxSin
```

```python
def approxSinGraph(max_x, step, max_terms):
    x_values = np.arange(0, max_x + 1e-12, step)
    exact_values = np.sin(x_values)
    
    errors = {}
    
    for terms in range (1, max_terms + 1):
        errors[terms] = []
        
    for x in x_values:
        exact = math.sin(x)
        for terms in range(1, max_terms + 1):
            Approx = taylorSin(x, terms)
            errors[terms].append(abs(Approx - exact))

    for terms in range(1, max_terms + 1):
        errors[terms] = np.array(errors[terms])

    return x_values, exact_values, errors
```

```python
x_values, exact_values, errors = approxSinGraph(10, 0.1, 4)
#values of x (1 to 10), step size, number of terms

print(len(x_values), len(errors[1]), len(errors[2]), len(errors[3]), len(errors[4]))

plt.figure()
plt.plot(x_values, errors[1], label="1st term")
plt.plot(x_values, errors[2], label="2 terms")
plt.plot(x_values, errors[3], label="3 terms")
plt.plot(x_values, errors[4], label="4 terms")
plt.yscale("log")
plt.xlabel("x")
plt.ylabel("absolute error")
plt.title("Trunication error for sin(x)")
plt.legend()
plt.show

# The error seems to decrease non-uniformly across the values of x.
```
