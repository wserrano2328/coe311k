```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler_init():
    h = 0.01 #step size
    x_start, x_end = 0, 2
    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1
    y_total_error = 0

    # Function representing dy/dx = -2y
    def f(x, y):
        return -2 * y

    # Implement Euler's method
    for i in range(num_steps):

        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
        y_true = np.exp(-2 * x_values[i+1])
        y_point_error = abs(y_true - y_values[i+1])
        y_total_error += y_point_error
        
    print("h =",h,"absolute error:", y_total_error, "\nError compared to first run: 0")
    
    # Plot the results

    plt.plot(x_values, y_values, label="Euler's Method Approximation")
    plt.xlabel('x')
    plt.ylabel('y')  
    plt.title("Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()
    plt.show()
    
    return y_values, y_total_error
```

```python
def euler(h, y_true_total):

    x_start, x_end = 0, 2
    num_steps = int((x_end - x_start) / h)


    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1
    y_total_error = 0

    # Function representing dy/dx = -2y
    def f(x, y):
        return -2 * y

    # Implement Euler's method
    for i in range(num_steps):

        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])  
        y_true = np.exp(-2 * x_values[i+1])
        
        y_point_error = abs(y_true - y_values[i+1])
        y_total_error += y_point_error
    
    y_comp_error = abs(y_total_error - y_true_total)
        
    
    print("h =",h,"absolute error:", y_total_error, "\nError compared to first run:",y_comp_error)
    
    # Plot the results

    plt.plot(x_values, y_values, label="Euler's Method Approximation")
    plt.xlabel('x')
    plt.ylabel('y')  
    plt.title("Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()
    plt.show()
```

```python
first_run_y, y_true_total = euler_init()

for i in np.arange(0.06, 1.06, 0.05):
    euler(i, y_true_total)
```
