```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 4        # Constant in the equation
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx

plt.figure(figsize=(10, 5))

for h in h_values:
    
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    plt.plot(x_values, y_values, label=f'h = {h}')

plt.axhline(0, color='black', linewidth=1, alpha=0.5)
plt.xlabel('x')
plt.ylabel('y (Displacement)')
plt.title(f"Stability Analysis of Euler's Method (k={k})")
plt.legend(title="Step Size (h)")
plt.grid(True)
plt.show()
```

```python
# We can see that when h = .2, it starts to get unstable around x = 6
#and will continue to become more unstable as x increases. 
# As x increases, all step sizes become more and more unstable, but when
#the step sizes increase, so does the rate at which the function becomes
#unstables.
# When increasing the k-value, the function oscillates faster, but as the
#k value increases, the error in approximation becomes more dramatic.
# Euler's method could not solve this with a larger h as it would be too 
#inaccurate to be able to determine the actual oscillation pattern that
#is happening.
```
