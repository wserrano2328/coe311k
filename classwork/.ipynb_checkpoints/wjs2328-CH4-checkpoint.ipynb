```python
import numpy as np
import math

def calculatePi(terms):
    ApproxPi = 0

    for i in range (0, terms):
        leibniz = 4 * (-1)**i / (2*i + 1)
        ApproxPi += leibniz
        
        if (i + 1) % 100 == 0:
            relativeError = abs(np.pi - ApproxPi) / abs(np.pi)
            print ("n={}, relative error={:.3e}".format(i+1, relativeError))
    
    error = abs(math.pi - ApproxPi)
    return ApproxPi, error

n_terms = 5000

approx, error = calculatePi(n_terms)





```
