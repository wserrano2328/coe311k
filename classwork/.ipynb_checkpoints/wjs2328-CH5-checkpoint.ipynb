```python
import numpy as np

def lu_decomposition(A):
    n = len(A)
    L = np.eye(n)
    U = np.array(A, dtype=float) #Copy's A to ensure it does not 
    #change the original matrix
    
    for i in range(n):
        
        for j in range(i+1, n):
            if U[i, i] == 0:
                raise valueError("Zero pivot encountered. LU decomposition cannont proceed.")
                
                factor = U[j, i] / U[i, j] #Multiplier for row operation
                
                L[j, i] = factor #Store multiplier in L
                
                for k in range(i, n):
                    U[j, k] -= factor * U[i, k]
                    
            return L, U
```

```python
A = np.array([[2, 3, 1],
              [4, 7, -1],
              [-2, 1, 2]])

L, U = lu_decomposition(A)

print("Matrix A: \n", A)
print("Lower Triangular Matrix L: \n", L)
print("Upper Triangular Matrix U: \n", U)
```

```python
import numpy as np

def gaussian_elimination(A, b):
    n = len(b)
    
    for i in range(n):
        max_row = max(range(i,n), key=lambda x: abs(A[x][i]))
        A[i], A[max_row] = A[max_row], A[i]
        b[i], b[max_row] = b[max_row], b[i]
        
        for j in range(i + 1, n):
            ratio = A[j][i] / A[i][i]
            A[j] = [A[j][k] - ratio * A[i][k] for k in range(n)]
            b[j] -= ratio * b[i]
            
    x = [0] * n
    for i in range (n - 1, -1, -1):
        x[i] = (b[i] - sum(A[i][j] * x[j] for j in range (i + 1,n))) / A[i][i]
        
        return x
```

```python
A = [[3, 2, -4], [2, 3, 3], [5, -3, 1]]

b = [3, 15, 14]

solution = gaussian_elimination(A, b)
print(f"Solution: {solution}")
```
