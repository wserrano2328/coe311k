```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

cs = CubicSpline(x, y, bc_type='natural')
x_inter = np.linspace(0, 5, 100)
y_inter = cs(x_inter)

coeffs = np.polyfit(x, y, deg=5)
poly = np.poly1d(coeffs)
y_poly = poly(x_inter)
```

```python
plt.figure(figsize=(9, 6))
plt.title("Cubic vs. Polynomial Fitting")
plt.plot(x, y, 'bo', markersize=8, label="Data points")
plt.plot(x_inter, y_inter, 'b', linewidth=1, label="Cubic Spline")
plt.plot(x_inter, y_poly, 'r', linewidth=2, label="Polynomial Fitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
