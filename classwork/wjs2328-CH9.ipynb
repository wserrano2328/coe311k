```python
import numpy as np
import matplotlib.pyplot as plt

#data points

X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

#Stack the design matric with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

#Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

#Had to change the print method for slope and intercept because jupyter was giving me issues

print("Slope:", m)
print("Intercept:", c)

Y_line = m*X + c

plt.scatter(X, Y, label="Data Points")
plt.plot(X, Y_line, label="Best-fit line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
rng = np.random.default_rng()

X_rand = rng.permutation(X)
Y_rand = rng.permutation(Y)

A_rand = np.vstack([X_rand, np.ones(len(X_rand))]).T

m_rand, c_rand = np.linalg.lstsq(A_rand, Y_rand, rcond=None)[0]

print("Slope:", m_rand)
print("Intercept:", c_rand)

Y_line_rand = m_rand*X_rand + c_rand

plt.scatter(X_rand, Y_rand, label="Random Points")
plt.plot(X_rand, Y_line_rand, label="Best-fit line")
plt.xlabel("X randomized")
plt.ylabel("Y randomized")
plt.legend()
plt.show()
```

```python
X3 = np.arange(0, 1.01, 0.01)

m_true = 2.5
c_true = 1.0

noise = rng.normal(0, 0.1, len(X3))

Y3 = m_true*X3 + c_true + noise

A3 = np.vstack([X3, np.ones(len(X3))]).T
m_fit, c_fit = np.linalg.lstsq(A3, Y3, rcond=None)[0]

print("True slope:", m_true)
print("True intercept:", c_true)
print("Estimated slope:", m_fit)
print("Estimated intercept:", c_fit)

Y_line3 = m_fit*X3 + c_fit

plt.scatter(X3, Y3, s=10, label="Bulk data")
plt.plot(X3, m_true*X3 + c_true, label="True line")
plt.plot(X3, Y_line3, label="Best-fit line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
