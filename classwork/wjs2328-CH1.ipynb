```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.subplot(2, 1, 1) # (row, column, active area)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'g')
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 1) # (row, column, active area)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(2, 2, 3)
y = np.sin(x)
plt.scatter(x,y)
plt.subplot(2, 2, 4)
plt.plot(x, np.tan(x), 'b')
plt.show()
```
