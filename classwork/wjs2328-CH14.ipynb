```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi 
N = 50

x = np.linspace(a, b, N)
y = f(x)

plt.plot(x, y, 'o-', label='Sampled Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Uniform Sampling of f(x) = sin(x)')
plt.legend()
plt.show()
```

```python
h = (b - a) / (N - 1)

df_dx_c = (f(x + h) - f(x - h)) / (2 * h)
df_dx_f = (f(x+h) - f(x)) / h
df_dx_b = (f(x) - (f(x-h))) / h

plt.plot(x, np.cos(x), label='Exact Derivative')
plt.plot(x, df_dx_b, 'g', label='Backward Difference Approximation')
plt.plot(x, df_dx_c, 'y', label='Central Difference Approximation')
plt.plot(x, df_dx_f, 'r', label='Forward Difference Approximation')
plt.plot()
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Approximating Derivative of f(x) = sin(x) using Central Difference")
plt.legend()
plt.show()
```

```python
#The backward difference is slightly ahead of the exact derivative while the forward difference is slight behind
#the exact derivative. The central difference approximation is the only line that is closest to reflecting the exact
#derivative.
```
