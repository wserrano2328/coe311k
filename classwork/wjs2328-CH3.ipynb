```python
import random
import numpy

def montecarlo_pi(num_samples):
    number_of_hits = 0

    for _ in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        if x * x + y * y < 1:
            number_of_hits = number_of_hits + 1

    return 4 * number_of_hits / num_samples
```

```python
import math

for i in range (0, 100):
    calculated_pi = montecarlo_pi(1000)
    true_pi = numpy.pi
    true_error = abs(true_pi - calculated_pi)
    relative_error = abs((true_pi - calculated_pi)/true_pi)
    
    print(calculated_pi, true_pi, relative_error)
```
